```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

# Read data from csv files
all_files = ['data/东方财富概念版块/水产养殖.csv']
dfs = []
for file in all_files:
    df = pd.read_csv(file)
    df['trade_date'] = pd.to_datetime(df['trade_date'], format='%Y%m%d')
    dfs.append(df)

# Combine all dataframes
all_data = pd.concat(dfs)

# Set font "SimHei"
font_path = "/Library/Fonts/SimHei.ttf"
font_prop = FontProperties(fname=font_path)

# Plot the trend of 'total_mv' over 'trade_date'
fig, ax = plt.subplots(figsize=(10, 6))

for i, df in enumerate(dfs):
    df.plot(kind='line', x='trade_date', y='total_mv', ax=ax, label=df['name'].iloc[0], color='C%d' % i)

ax.xaxis.set_major_formatter(DateFormatter('%Y-%m-%d'))
ax.tick_params(axis='y', labelsize=12)
ax.tick_params(axis='x', labelsize=12)
ax.legend(prop=font_prop, loc='upper left')
ax.set_xlabel('Trade Date', fontproperties=font_prop)
ax.set_ylabel('Total MV (万元)', fontproperties=font_prop)
ax.set_title('水产养殖行业研究报告', fontproperties=font_prop)

# Save the plot
fig.savefig('/Users/ash/Desktop/本科/毕业论文/writer/output/2025水产养殖行业研究报告/image/东方财富概念版块.png')
```
