```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import os

# Load the data
file_path = '/Users/ash/Desktop//writer/data//.csv'
df = pd.read_csv(file_path)

# Make a copy of the dataframe for processing
processed_df = df.copy()

# Convert trade_date to datetime format
processed_df['trade_date'] = pd.to_datetime(processed_df['trade_date'], format='%Y%m%d')

# Extract time-related features
processed_df['year'] = processed_df['trade_date'].dt.year
processed_df['month'] = processed_df['trade_date'].dt.month
processed_df['day'] = processed_df['trade_date'].dt.day

# Set up plot
font_path = "/Library/Fonts/SimHei.ttf"
font_prop = FontProperties(fname=font_path)
fig, ax = plt.subplots(figsize=(12, 6))

# Plot total_mv over time
ax.plot(processed_df['trade_date'], processed_df['total_mv'], label='')

# Format plot
ax.set_xlabel('', fontproperties=font_prop)
ax.set_ylabel(' ()', fontproperties=font_prop)
ax.set_title(' - ', fontproperties=font_prop)
ax.legend(prop=font_prop)

# Create output directory if it doesn't exist
output_dir = '/Users/ash/Desktop//writer/output/2025/image'
os.makedirs(output_dir, exist_ok=True)

# Save plot
output_path = os.path.join(output_dir, '_.png')
fig.savefig(output_path, bbox_inches='tight')
plt.close()

# Show the processed dataframe with new features
processed_df.head()
```
