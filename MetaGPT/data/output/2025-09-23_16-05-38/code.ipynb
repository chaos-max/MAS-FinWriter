```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# Load the dataset
data1 = pd.read_csv('data/申万行业指数/水产养殖.csv')
data2 = pd.read_csv('data/申万行业指数/渔业.csv')

# Convert 'trade_date' to datetime format
data1['trade_date'] = pd.to_datetime(data1['trade_date'], format='%Y%m%d')
data2['trade_date'] = pd.to_datetime(data2['trade_date'], format='%Y%m%d')

# Extract year, month, and day from 'trade_date'
data1['Year'] = data1['trade_date'].dt.year
data1['Month'] = data1['trade_date'].dt.month
data1['Day'] = data1['trade_date'].dt.day

data2['Year'] = data2['trade_date'].dt.year
data2['Month'] = data2['trade_date'].dt.month
data2['Day'] = data2['trade_date'].dt.day

# Plot the trend of 'total_mv' over 'trade_date' for both files
font_path = "/Library/Fonts/SimHei.ttf"
font = FontProperties(fname=font_path)

fig, ax = plt.subplots()

ax.plot(data1['trade_date'], data1['total_mv'], label='水产养殖', color='blue')
ax.plot(data2['trade_date'], data2['total_mv'], label='渔业', color='red')

ax.set_xlabel('Trade Date')
ax.set_ylabel('Total MV (万元)')
ax.set_title('Total MV over Time')
ax.set_font_name(font.get_name())
ax.legend(prop=font)

# Save the plot
fig.savefig('/Users/ash/Desktop/本科/毕业论文/writer/output/2025水产养殖行业研究报告/image/申万行业指数.png')
```
