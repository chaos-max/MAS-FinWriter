```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import os

# Define font
font_path = "/Library/Fonts/SimHei.ttf"
font_prop = FontProperties(fname=font_path)

# List of files to process
file_paths = [
    '/Users/ash/Desktop//writer/data//.csv',
    '/Users/ash/Desktop//writer/data//.csv',
    '/Users/ash/Desktop//writer/data//.csv'
]

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(12, 6))

# Process each file
for file_path in file_paths:
    # Read the data
    df = pd.read_csv(file_path)
    
    # Convert trade_date to datetime
    df['trade_date'] = pd.to_datetime(df['trade_date'], format='%Y%m%d')
    
    # Extract time-related features
    df['year'] = df['trade_date'].dt.year
    df['month'] = df['trade_date'].dt.month
    df['day'] = df['trade_date'].dt.day
    
    # Get filename for legend
    file_name = os.path.basename(file_path).replace('.csv', '')
    
    # Plot total_mv over trade_date
    ax.plot(df['trade_date'], df['total_mv'], label=file_name)

# Set plot properties
ax.set_xlabel('', fontproperties=font_prop)
ax.set_ylabel(' ()', fontproperties=font_prop)
ax.set_title('', fontproperties=font_prop)
ax.legend(prop=font_prop)
ax.grid(True)

# Create output directory if not exists
output_dir = '/Users/ash/Desktop//writer/output/25/image'
os.makedirs(output_dir, exist_ok=True)

# Save the figure
output_path = os.path.join(output_dir, '.png')
fig.savefig(output_path, bbox_inches='tight', dpi=300)

# Show the plot
plt.show()
```

```
# Create subplots for each financial metric
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('', fontproperties=font_prop)

# Plot each metric in separate subplot
metrics = ['total_mv', 'pe', 'pb']
titles = [' ()', ' (PE)', ' (PB)']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

for i, (metric, title) in enumerate(zip(metrics, titles)):
    ax = axes[i//2, i%2]
    for file_path, color in zip(file_paths, colors):
        df = pd.read_csv(file_path)
        df['trade_date'] = pd.to_datetime(df['trade_date'], format='%Y%m%d')
        file_name = os.path.basename(file_path).replace('.csv', '')
        ax.plot(df['trade_date'], df[metric], label=file_name, color=color, linewidth=2)
    
    ax.set_xlabel('', fontproperties=font_prop)
    ax.set_ylabel(title, fontproperties=font_prop)
    ax.legend(prop=font_prop)
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.tick_params(axis='x', rotation=45)

# Remove empty subplot if odd number of metrics
if len(metrics) % 2 != 0:
    fig.delaxes(axes[-1, -1])

# Adjust layout and save
plt.tight_layout()
output_path = os.path.join(output_dir, '.png')
fig.savefig(output_path, bbox_inches='tight', dpi=300)
plt.show()
```
