```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import os

# Define file paths
file_paths = [
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/动力煤.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/焦炭Ⅱ.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/焦炭Ⅲ.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/焦煤.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/煤化工.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/煤炭.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/煤炭开采.csv'
]

# Load and preprocess each file
data_frames = []
for file_path in file_paths:
    # Read CSV file
    df = pd.read_csv(file_path)
    
    # Make a copy of the dataframe
    df_processed = df.copy()
    
    # Convert trade_date to datetime format
    df_processed['trade_date'] = pd.to_datetime(df_processed['trade_date'], format='%Y%m%d')
    
    # Extract time-related features
    df_processed['year'] = df_processed['trade_date'].dt.year
    df_processed['month'] = df_processed['trade_date'].dt.month
    df_processed['day'] = df_processed['trade_date'].dt.day
    
    # Add filename as identifier
    df_processed['file_name'] = os.path.basename(file_path).replace('.csv', ')
    
    data_frames.append(df_processed)

# Combine all dataframes for plotting
combined_df = pd.concat(data_frames)

# Set up plot
font_path = "/Library/Fonts/SimHei.ttf"
font_prop = FontProperties(fname=font_path)
fig, ax = plt.subplots(figsize=(12, 6))

# Plot each file's data
for name, group in combined_df.groupby('file_name'):
    ax.plot(group['trade_date'], group['total_mv'], label=name)

# Customize plot
ax.set_xlabel('交易日期', fontproperties=font_prop)
ax.set_ylabel('总市值 (万元)', fontproperties=font_prop)
ax.set_title('申万行业指数 - 煤炭相关行业总市值趋势', fontproperties=font_prop)
ax.legend(prop=font_prop)
plt.xticks(rotation=45)

# Save plot
output_dir = '/Users/ash/Desktop/毕业/writer/output/煤炭行业研究报告_2023/image'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, '申万行业指数_煤炭相关行业总市值趋势.png')
fig.savefig(output_path, bbox_inches='tight', dpi=300)

# Show plot
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import os

# Load the data
file_path = '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/煤化工.csv'
data = pd.read_csv(file_path)

# Make a copy of the dataframe for processing
processed_data = data.copy()

# Convert trade_date to datetime format
processed_data['trade_date'] = pd.to_datetime(processed_data['trade_date'], format='%Y%m%d')

# Extract time-related features
processed_data['year'] = processed_data['trade_date'].dt.year
processed_data['month'] = processed_data['trade_date'].dt.month
processed_data['day'] = processed_data['trade_date'].dt.day

# Set up the plot
font_path = "/Library/Fonts/SimHei.ttf"
font_prop = FontProperties(fname=font_path)

fig, ax = plt.subplots(figsize=(12, 6))

# Plot total_mv over time
ax.plot(processed_data['trade_date'], processed_data['total_mv'], 
        label='煤化工', color='blue')

# Format the plot
ax.set_xlabel('交易日期', fontproperties=font_prop)
ax.set_ylabel('总市值 (万元)', fontproperties=font_prop)
ax.set_title('东方财富概念版块 - 总市值趋势', fontproperties=font_prop)
ax.legend(prop=font_prop)

# Ensure the output directory exists
output_dir = '/Users/ash/Desktop/毕业/writer/output/煤炭行业研究报告_2023/image'
os.makedirs(output_dir, exist_ok=True)

# Save the plot
output_path = os.path.join(output_dir, '东方财富概念版块_总市值趋势.png')
fig.savefig(output_path, bbox_inches='tight', dpi=300)

# Show the processed data head
processed_data.head()
```

```
# Get all files in the directory (assuming there might be multiple files)
data_dir = '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块'
all_files = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith('.csv')]

# Create figure and set up font
font_prop = FontProperties(fname="/Library/Fonts/SimHei.ttf")
fig, ax = plt.subplots(figsize=(14, 8))

# Plot each file's data
for file_path in all_files:
    # Read and process data
    data = pd.read_csv(file_path)
    data['trade_date'] = pd.to_datetime(data['trade_date'], format='%Y%m%d')
    
    # Get the base filename for legend
    file_name = os.path.basename(file_path).replace('.csv', '')
    
    # Plot total_mv (convert to 万元 by dividing by 10000)
    ax.plot(data['trade_date'], data['total_mv']/10000, 
            label=file_name, 
            marker='o', markersize=3, linestyle='-', linewidth=1)

# Format plot
ax.set_xlabel('交易日期', fontproperties=font_prop)
ax.set_ylabel('总市值 (万元)', fontproperties=font_prop)
ax.set_title('东方财富概念版块 - 总市值趋势比较', fontproperties=font_prop)
ax.legend(prop=font_prop)
ax.grid(True, linestyle='--', alpha=0.6)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Ensure output directory exists and save plot
output_dir = '/Users/ash/Desktop/毕业/writer/output/煤炭行业研究报告_2023/image'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, '东方财富概念版块_总市值趋势比较.png')
fig.savefig(output_path, bbox_inches='tight', dpi=300)

plt.show()
```
