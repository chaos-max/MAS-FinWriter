```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# Load the dataset
data1 = pd.read_csv('data//.csv')
data2 = pd.read_csv('data//.csv')

# Convert trade_date to datetime format
data1['trade_date'] = pd.to_datetime(data1['trade_date'], format='%Y%m%d')
data2['trade_date'] = pd.to_datetime(data2['trade_date'], format='%Y%m%d')

# Extract year, month, and day from trade_date
data1['Year'] = data1['trade_date'].dt.year
data1['Month'] = data1['trade_date'].dt.month
data1['Day'] = data1['trade_date'].dt.day

data2['Year'] = data2['trade_date'].dt.year
data2['Month'] = data2['trade_date'].dt.month
data2['Day'] = data2['trade_date'].dt.day

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the trend of relevant financial metrics over trade_date for all files on the same graph
ax.plot(data1['trade_date'], data1['total_mv'], label='', color='blue')
ax.plot(data2['trade_date'], data2['total_mv'], label='', color='red')

# Set font "SimHei"
font_path = "/Library/Fonts/SimHei.ttf"
font = FontProperties(fname=font_path)

# Set legend
ax.legend(prop=font)

# Set x and y axis labels
ax.set_xlabel('Trade Date', fontproperties=font)
ax.set_ylabel('Total MV ()', fontproperties=font)

# Set title
ax.set_title(' - ', fontproperties=font)

# Save the plot
fig.savefig('/Users/ash/Desktop///writer/output/2025/image/.png')
```

```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

data_path = 'data//.csv'
df = pd.read_csv(data_path)
df['trade_date'] = pd.to_datetime(df['trade_date'], format='%Y%m%d')

# Extract time-related features
df['Year'] = df['trade_date'].dt.year
df['Month'] = df['trade_date'].dt.month
df['Day'] = df['trade_date'].dt.day

# Set font
font_path = "/Library/Fonts/SimHei.ttf"
font = FontProperties(fname=font_path)

# Plot total_mv over trade_date for all files
plt.figure(figsize=(10,6))
for i, filename in enumerate(df['name'].unique()):
    df_file = df[df['name'] == filename]
    plt.plot(df_file['trade_date'], df_file['total_mv']*1e4, label=filename, marker='o')  # convert to 10,000
plt.xlabel('Trade Date', fontproperties=font)
plt.ylabel('Total MV ()', fontproperties=font)
plt.title('Trend of Total MV over Trade Date', fontproperties=font)
plt.legend(prop=font)
plt.grid(True)

# Save the plot
output_path = '/Users/ash/Desktop///writer/output/2025/image'
plt.savefig(output_path + '/_.png')
```
