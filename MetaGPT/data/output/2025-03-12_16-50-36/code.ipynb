```
import pandas as pd
import matplotlib.pyplot as plt
import os
from datetime import datetime

# List of file paths
file_paths = [
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/AI制药.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/AI手机.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/AI智能体.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/AI眼镜.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/AI芯片.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/AI语料.csv'
]

# Function to preprocess each file
def preprocess_file(file_path):
    df = pd.read_csv(file_path)
    # Convert 'trade_date' to datetime format
    df['trade_date'] = pd.to_datetime(df['trade_date'], format='%Y%m%d')
    # Extract year, month, and day
    df['year'] = df['trade_date'].dt.year
    df['month'] = df['trade_date'].dt.month
    df['day'] = df['trade_date'].dt.day
    return df

# Preprocess all files
preprocessed_data = [preprocess_file(file_path) for file_path in file_paths]

# Plotting the trend of 'total_mv' over 'trade_date' for all files
plt.figure(figsize=(14, 7))
for i, df in enumerate(preprocessed_data):
    plt.plot(df['trade_date'], df['total_mv'], label=os.path.basename(file_paths[i]).replace('.csv', ''))

# Setting up the plot
plt.xlabel('交易日期', fontproperties='SimHei')
plt.ylabel('总市值', fontproperties='SimHei')
plt.title('东方财富概念版块总市值趋势', fontproperties='SimHei')
plt.legend(prop={'family': 'SimHei'})
plt.grid(True)

# Save the plot
output_dir = '/Users/ash/Desktop/毕业/writer/output/AI研究/image'
os.makedirs(output_dir, exist_ok=True)
plt.savefig(os.path.join(output_dir, '东方财富概念版块_总市值趋势.png'))

# Display the plot
plt.show()
```
