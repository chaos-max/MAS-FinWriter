```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import os

# Define file paths
file_paths = [
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/半导体.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/半导体材料.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/半导体设备.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/分立器件.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/集成电路封测.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/数字芯片设计.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/模拟芯片设计.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/被动元件.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/电子化学品Ⅱ.csv',
    '/Users/ash/Desktop/毕业/writer/data/申万行业指数/电子化学品Ⅲ.csv'
]

# Load and preprocess all files
dfs = []
for file_path in file_paths:
    df = pd.read_csv(file_path)
    # Convert trade_date to datetime
    df['trade_date'] = pd.to_datetime(df['trade_date'], format='%Y%m%d')
    # Extract time features
    df['year'] = df['trade_date'].dt.year
    df['month'] = df['trade_date'].dt.month
    df['day'] = df['trade_date'].dt.day
    # Add filename as identifier
    df['file_name'] = os.path.basename(file_path).replace('.csv', '')
    dfs.append(df)

# Create plot
font_path = "/Library/Fonts/SimHei.ttf"
font_prop = FontProperties(fname=font_path)

fig, ax = plt.subplots(figsize=(12, 8))

# Plot each dataframe
for df in dfs:
    ax.plot(df['trade_date'], df['total_mv'], label=df['file_name'].iloc[0])

# Set plot properties
ax.set_xlabel('交易日期', fontproperties=font_prop)
ax.set_ylabel('总市值 (万元)', fontproperties=font_prop)
ax.set_title('申万行业指数总市值趋势', fontproperties=font_prop)
ax.legend(prop=font_prop)
plt.xticks(rotation=45)

# Save plot
output_dir = '/Users/ash/Desktop/毕业/writer/output/25_SemiRpt/image'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, '申万行业指数总市值趋势.png')
fig.savefig(output_path, bbox_inches='tight', dpi=300)
plt.close()
```

```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import os

# Define the file paths
file_paths = [
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/AI芯片.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/Chiplet概念.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/EDA概念.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/IGBT概念.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/半导体概念.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/国产芯片.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/汽车芯片.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/第三代半导体.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/第四代半导体.csv',
    '/Users/ash/Desktop/毕业/writer/data/东方财富概念版块/存储芯片.csv'
]

# Create a dictionary to store all dataframes
dataframes = {}

# Process each file
for file_path in file_paths:
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Make a copy of the dataframe for processing
    processed_df = df.copy()
    
    # Convert trade_date to datetime format
    processed_df['trade_date'] = pd.to_datetime(processed_df['trade_date'], format='%Y%m%d')
    
    # Extract year, month, and day features
    processed_df['year'] = processed_df['trade_date'].dt.year
    processed_df['month'] = processed_df['trade_date'].dt.month
    processed_df['day'] = processed_df['trade_date'].dt.day
    
    # Store the processed dataframe using the filename as key
    file_name = os.path.basename(file_path).replace('.csv', '')
    dataframes[file_name] = processed_df

# Verify the first dataframe's structure
print(f"Processed {len(dataframes)} files. First file structure:")
print(dataframes[next(iter(dataframes))].head())
```
