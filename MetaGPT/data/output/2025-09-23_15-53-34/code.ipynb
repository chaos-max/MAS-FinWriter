```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# Load the dataset
data1 = pd.read_csv('data/申万行业指数/水产养殖.csv')
data2 = pd.read_csv('data/申万行业指数/渔业.csv')

# Convert trade_date to datetime format
data1['trade_date'] = pd.to_datetime(data1['trade_date'], format='%Y%m%d')
data2['trade_date'] = pd.to_datetime(data2['trade_date'], format='%Y%m%d')

# Extract year, month, and day from trade_date
data1['Year'] = data1['trade_date'].dt.year
data1['Month'] = data1['trade_date'].dt.month
data1['Day'] = data1['trade_date'].dt.day

data2['Year'] = data2['trade_date'].dt.year
data2['Month'] = data2['trade_date'].dt.month
data2['Day'] = data2['trade_date'].dt.day

# Plot the trend of total_mv over trade_date for both files
fig, ax = plt.subplots()

# Set font
font_path = "/Library/Fonts/SimHei.ttf"
font = FontProperties(fname=font_path)

ax.plot(data1['trade_date'], data1['total_mv'], label='水产养殖', color='red')
ax.plot(data2['trade_date'], data2['total_mv'], label='渔业', color='blue')

ax.set_xlabel('Trade Date', fontproperties=font)
ax.set_ylabel('Total MV (万元)', fontproperties=font)
ax.set_title('Trend of Total MV Over Time', fontproperties=font)
ax.legend(loc='upper left')

# Save the plot
fig.savefig('/Users/ash/Desktop/本科/毕业论文/writer/output/2025水产养殖行业研究报告/image/申万行业指数.png')
```

```
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# Set font
font_path = "/Library/Fonts/SimHei.ttf"
font = FontProperties(fname=font_path)

# Read data
data_path = 'data/东方财富概念版块/水产养殖.csv'
df = pd.read_csv(data_path)
df['trade_date'] = pd.to_datetime(df['trade_date'], format='%Y%m%d')

# Extract time-related features
df['Year'] = df['trade_date'].dt.year
df['Month'] = df['trade_date'].dt.month
df['Day'] = df['trade_date'].dt.day

# Plot
plt.figure(figsize=(12, 6))
plt.gca().xaxis_date()
plt.gca().xaxis.set_major_formatter(plt.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(plt.DateLocator(interval=pd.DateOffset(months=1)))
plt.gca().xaxis.set_minor_locator(plt.DateLocator(interval=pd.DateOffset(days=1)))
plt.gca().xaxis.set_minor_formatter(plt.DateFormatter('%d'))
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: format(x, ',.2f')))
plt.gca().yaxis.set_major_locator(plt.MaxNLocator(integer=True))
plt.gca().yaxis.set_ticks_position('none')
plt.gca().xaxis.set_ticks_position('both')
plt.gca().grid(True)

colors = ['red', 'green', 'blue', 'yellow']
styles = ['-', '--', '-.', ':']

for i, (file, color, style) in enumerate(zip(df['name'], colors, styles)):
    plt.plot(df[df['name'] == file]['trade_date'], df[df['name'] == file]['total_mv'], color=color, linestyle=style, label=file)

plt.legend(prop=font)
plt.title('水产养殖行业研究报告', prop=font)
plt.xlabel('日期', prop=font)
plt.ylabel('总市值（万元）', prop=font)

plt.subplots_adjust(left=0.1, bottom=0.25)

# Save the plot
output_path = '/Users/ash/Desktop/本科/毕业论文/writer/output/2025水产养殖行业研究报告/image'
plt.savefig(output_path + '/东方财富概念版块_水产养殖行业研究报告.png')
```
